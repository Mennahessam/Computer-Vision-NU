```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
img = data.camera()
```

```python
plt.imshow(img,cmap='gray')
```

```python
def apply_kernel(image, kernel):
    rows, cols = image.shape[:2]
    k_rows, k_cols = kernel.shape[:2]
    
    filtered_image = np.zeros_like(image)
    
    for i in range(rows - k_rows + 1):
        for j in range(cols - k_cols + 1):
            window = image[i:i+k_rows, j:j+k_cols]
            filtered_pixel = (kernel * window).sum()
            filtered_image[i+k_rows//2, j+k_cols//2] = filtered_pixel
    
    return filtered_image
```

```python
kernel = np.array(
[[-2, -1,  0],
 [-1,  1,  1],
 [ 0,  1,  2]]
)
plt.imshow(apply_kernel(img,kernel),cmap='gray')
```
